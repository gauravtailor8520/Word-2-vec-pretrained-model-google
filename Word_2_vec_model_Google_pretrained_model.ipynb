```python
import gensim
from gensim.models import Word2Vec,KeyedVectors
```

Pretrainde model link : - https://drive.google.com/file/d/1ETEzH8X7uM_xXtIEuNLgz9VL7eQEeE_V/view?usp=sharing

```python
# Load Google's pre-trained Word2Vec model.
model = KeyedVectors.load_word2vec_format('/content/drive/MyDrive/NLP/GoogleNews-vectors-negative300.bin.gz', binary=True)
```

```python
model["man"]
```

```python
vec = model['INR'] - model ['India'] + model['England']
model.most_similar([vec])
```

```python
model.similarity('man','woman')
```

```python
model.similarity('man','PHP')
```

```python
model.doesnt_match(['PHP','java','monkey'])
```

```python
vec = model['king'] - model['man'] + model['woman']
model.most_similar([vec])
```
